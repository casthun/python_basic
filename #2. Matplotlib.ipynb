```python
df = pd.read_excel("./sample_data/concat.xlsx")
df.columns
```

```python
df1 = df[["", "", ""]]
df1
```

```python
df2 = df1[''].value_counts().reset_index()
df2.columns = ['', '']
df2.sort_values(by=[''], ascending=True, inplace=True)
df2
```

```python
df3 = df1.groupby("")[""].sum().reset_index()
df3
```

```python
df_merge = pd.merge(df2, df3, on="")
df_merge
```

```python
#   

plt.scatter(df_merge[''], df_merge[''], color='b', alpha=0.5)

font2 = {"family": "AppleGothic",
        'color': 'b',
        'weight': "bold",
        'size': 16}

plt.xlabel("", fontdict=font2)
plt.ylabel("", fontdict=font2)

plt.title(" R&D ", fontdict=font2, pad=20)


plt.show()
```

```python
#     

import numpy as np

plt.scatter(df_merge[''], df_merge[''], c=np.random.rand(len(df_merge[''])))

font2 = {"family": "AppleGothic",
        'color': 'b',
        'weight': "bold",
        'size': 20}

plt.xlabel("", fontdict=font2)
plt.ylabel("", fontdict=font2)

plt.title(" R&D ", fontdict=font2, pad=20)

# for plt.text()    () 
for i, j, k in zip(df_merge[''], df_merge[''], df_merge['']):
    plt.text(i, j, k, rotation=50)

plt.show()
```

```python
import pandas as pd

df = pd.read_excel("./sample_data/concat.xlsx")
df
```

```python
df1 = df[['', '']]
df1
```

```python
dic_gov = {"":"",
           "":"",
           "":"",
           "": "",
           "":"",
           "": "",
           "":""}
```

```python
gov = lambda x: dic_gov.get(x, x)
```

```python
df1[''] = df1[''].map(gov)
df1
```

```python
df1 = df1.groupby("")[''].sum().reset_index()
df1
```

```python
df1[''] = df1['']/df1[''].sum()*100
df1
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"
plt.pie(df1[''], labels=df1[''], autopct="%.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"
plt.pie(df1[''], labels=df1[''], autopct="%.2f%%", explode=[0.05]*len(df1))
plt.show()
```
