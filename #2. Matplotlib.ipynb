```python
plt.rcParams['font.family'] = "AppleGothic"
plt.plot(df_year['기준년도'], df_year['과제건수'], 'ro--')
plt.bar(df_year['기준년도'], df_year['과제건수'], color='y')
plt.xticks(df_year['기준년도'], rotation=45)
plt.title("년도별 과제건수", weight='bold')
plt.xlabel("기준년도")
plt.ylabel("과제건수")
plt.show()
```

```python
# 범례 표시
# plt.plot(), plt.bar() 함수 안에 label 인자를 이용하여 범례 표시

plt.rcParams['font.family'] = "AppleGothic"
plt.plot(df_year['기준년도'], df_year['과제건수'], 'ro-.', label="선 그래프")
plt.bar(df_year['기준년도'], df_year['과제건수'], label="막대 그래프")
plt.legend()

font1 = {"family": "AppleGothic",
        'color': 'b',
        'weight': 'bold',
        'size': 25}

font2 = {"family": "AppleGothic",
        'color': 'deeppink',
        'weight': "normal",
        'size': 24}

plt.xticks(list(df_year['기준년도']), rotation=45, font="Arial")
plt.yticks(font="Georgia")

plt.xlabel("Year", fontdict = font1)
plt.ylabel("과제건수", fontdict = font2)

plt.tick_params(axis='x', labelcolor="b", labelsize=12)
plt.tick_params(axis='y', labelcolor='deeppink', labelsize=12)

plt.title("년도별 과제건수")
plt.show()
```

```python
df = pd.read_excel("./sample_data/concat.xlsx")
df.columns
```

```python
df1 = df[["기준년도", "부처명", "정부투자연구비"]]
df1
```

```python
df2 = df1['기준년도'].value_counts().reset_index()
df2.columns = ['기준년도', '과제건수']
df2.sort_values(by=['기준년도'], ascending=True, inplace=True)
df2
```

```python
df3 = df1.groupby("기준년도")["정부투자연구비"].sum().reset_index()
df3
```

```python
df_merge = pd.merge(df2, df3, on="기준년도")
df_merge
```

```python
# 산점도 그래프 그리기

plt.scatter(df_merge['과제건수'], df_merge['정부투자연구비'], color='b', alpha=0.5)

font2 = {"family": "AppleGothic",
        'color': 'b',
        'weight': "bold",
        'size': 16}

plt.xlabel("과제건수", fontdict=font2)
plt.ylabel("정부투자연구비", fontdict=font2)

plt.title("연도별 R&D 투자현황", fontdict=font2, pad=20)


plt.show()
```

```python
# 산점도 데이터별 해당년도 그래프에 표시

import numpy as np

plt.scatter(df_merge['과제건수'], df_merge['정부투자연구비'], c=np.random.rand(len(df_merge['과제건수'])))

font2 = {"family": "AppleGothic",
        'color': 'b',
        'weight': "bold",
        'size': 20}

plt.xlabel("과제건수", fontdict=font2)
plt.ylabel("정부투자연구비", fontdict=font2)

plt.title("연도별 R&D 투자현황", fontdict=font2, pad=20)

# for문과 plt.text() 함수를 이용하여 마커 라벨(기준년도) 생성
for i, j, k in zip(df_merge['과제건수'], df_merge['정부투자연구비'], df_merge['기준년도']):
    plt.text(i, j, k, rotation=50)

plt.show()
```

```python
import pandas as pd

df = pd.read_excel("./sample_data/concat.xlsx")
df
```

```python
df1 = df[['부처명', '정부투자연구비']]
df1
```

```python
dic_gov = {"산업자원부":"산업통상자원부",
           "교육과학기술부":"과학기술정보통신부",
           "미래창조과학부":"과학기술정보통신부",
           "농촌진흥청": "농림축산식품부",
           "교육인적자원부":"교육부",
           "중소기업청": "중소벤처기업부",
           "지식경제부":"기획재정부"}
```

```python
gov = lambda x: dic_gov.get(x, x)
```

```python
df1['부처명'] = df1['부처명'].map(gov)
df1
```

```python
df1 = df1.groupby("부처명")['정부투자연구비'].sum().reset_index()
df1
```

```python
df1['비중'] = df1['정부투자연구비']/df1['정부투자연구비'].sum()*100
df1
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"
plt.pie(df1['비중'], labels=df1['부처명'], autopct="%.1f%%")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = "AppleGothic"
plt.pie(df1['비중'], labels=df1['부처명'], autopct="%.2f%%", explode=[0.05]*len(df1))
plt.show()
```

```python
import numpy as np

lst_inst = ['대기업', '중견기업', '중소기업', '대학', '출연연구소', '국공립연구소', '기타']

df = pd.read_excel("./sample_data/concat.xlsx")
df1 = df[['부처명', '연구수행주체', '기술수명주기', '연구개발단계', '정부투자연구비']]

idx = np.arange(1, 8)
a = []

for i in lst_inst:
    df_temp = df1[df1['연구수행주체']==i]
    a.append(np.array(df_temp['정부투자연구비']))
    
plt.rcParams['font.family'] = "AppleGothic"
plt.boxplot(a)
plt.xticks(idx, lst_inst)
plt.title("연구수행주체별 정부투자연구비 현황")
plt.show()
```
