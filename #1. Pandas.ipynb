```python
df1 = df.groupby('부처명')['정부투자연구비'].sum()
df1 = df1.reset_index()
df1
```

```python
# 부처명 수정1

dic_gov = {"산업자원부":"산업통상자원부",
           "교육과학기술부":"과학기술정보통신부",
           "미래창조과학부":"과학기술정보통신부",
           "농촌진흥청": "농림축산식품부",
           "교육인적자원부":"교육부",
           "중소기업청": "중소벤처기업부",
           "지식경제부":"기획재정부"}

for i, j in enumerate(df1['부처명']):
    if j not in dic_gov:
        pass
    else:
        df1.loc[i, '부처명'] = dic_gov[j]
df1
```

```python
# 부처명 수정2

dic_gov = {"산업자원부":"산업통상자원부",
           "교육과학기술부":"과학기술정보통신부",
           "미래창조과학부":"과학기술정보통신부",
           "농촌진흥청": "농림축산식품부",
           "교육인적자원부":"교육부",
           "중소기업청": "중소벤처기업부",
           "지식경제부":"기획재정부"}

for i, j in enumerate(df1['부처명']):
    df1.loc[i, '부처명'] = dic_gov.get(j, j)
    
df1
```

```python
# 부처명 수정3

dic_gov = {"산업자원부":"산업통상자원부",
           "교육과학기술부":"과학기술정보통신부",
           "미래창조과학부":"과학기술정보통신부",
           "농촌진흥청": "농림축산식품부",
           "교육인적자원부":"교육부",
           "중소기업청": "중소벤처기업부",
           "지식경제부":"기획재정부"}

gov = lambda x: dic_gov.get(x, x)

df1['부처명'] = df1['부처명'].map(gov)

df1
```

```python
df2 = df1.groupby('부처명')['정부투자연구비'].sum().sort_values(ascending=False)
df2 = df2.reset_index()
df2
```

```python
df1 = df['부처명'].value_counts().reset_index()
df1.columns = ["부처명", "과제건수"]
df1['부처명'] = df1['부처명'].map(gov)
df1 = df1.groupby('부처명')['과제건수'].sum()
df1 = df1.sort_values(ascending=False).reset_index()
df1
```

```python
df2 = df.groupby('부처명')["정부투자연구비"].sum().reset_index()
df2['부처명'] = df2['부처명'].map(gov)
df2 = df2.groupby('부처명')["정부투자연구비"].sum().sort_values(ascending=False).reset_index()
df2
```

```python
df_merge = pd.merge(df1, df2, on='부처명')
df_merge
```

```python
df_merge.to_excel("./sample_data/project_merge.xlsx", index=False)
```
